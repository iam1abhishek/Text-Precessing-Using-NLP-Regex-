Sample text

```python
#give your text according to your need
text="In 1996 Jimmy Wales, a successful bond trader, moved to San Diego, California, to establish Bomis, Inc., a Web portal company. In March 2000 Wales founded Nupedia, a free online encyclopaedia, with Larry Sanger as editor in chief. Nupedia was organized like existing encyclopaedias, with an advisory board of experts and a lengthy review process. By January 2001 fewer than two dozen articles were finished, and Sanger advocated supplementing Nupedia with an open-source encyclopaedia based on wiki software. On January 15, 2001, Wikipedia was launched as a feature of Nupedia.com, but, following objections from the advisory board, it was relaunched as an independent Web site a few days later. In its first year Wikipedia expanded to some 20,000 articles in 18 languages, including French, German, Polish, Dutch, Hebrew, Chinese, and Esperanto. In 2003 Nupedia was terminated and its articles moved into Wikipedia."
```

# Basic Regex Operation
Regex allows you to perform various operation on text such as pattern, replacing text and splitting text.

# Searching for a pattern

```python
pattern=r"encyclopaedia"
matches=re.findall(pattern,text)
print("Matches:" ,matches)
```

# Replacing text

```python
pattern=r"encyclopaedia"
replacement="reference work"
new_text=re.sub(pattern,replacement,text)
print("New_text", new_text)
```

Again finding the word "encyclopaedia"

```python
pattern=r"encyclopaedia"
matches=re.findall(pattern,new_text)
print("Matches:" ,matches)
```

#Splitting the words

```python
pattern=r"\s" #split on spaces
split_text=re.split(pattern,text)
print("Split Text:",split_text)
```

# Charcters Classes and Quantifiers

```python
#Characters classes
pattern=r"[aeiou]"
vowels=re.findall(pattern,text)
print("Vowels:",vowels)
```

**Quantifiers**

```python
pattern=r"e{2}" #matches two ansecutive 'e' characters
matches=re.findall(pattern,text)
print("Matches:",matches)
```

# Metacharacters and Anchors

```python
pattern=r"\b\w{3}\b" #matches 3 letter words
matches=re.findall(pattern,text)
print("Matches:",matches)
```

Anchors

```python
pattern=r"^the" #Matches 'the' at the begining of the string
matches=re.findall(pattern,text)
print("Matches:",matches)
```

#grouping and Capturing

```python
pattern=r"(\w+)\s(\w+)"  #matches two consecutive words
matches=re.findall(pattern,text)
print("Matches:",matches)
```

#Using Rogex for Tokenization
regex is also used for tokenization text into words and sentences.

# Tokenizing into Words

```python
pattern=r"\s" #split on spaces
words=re.split(pattern,text)
print("words:",words)
```

#Tokenization into Senetnces

```python
pattern=r"\.\s"  #split after followed by a space
sentences=re.split(pattern,text)
print("Sentences:",sentences)
```
